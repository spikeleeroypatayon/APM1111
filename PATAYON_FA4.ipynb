### Find the (a) first, (b) second, (c) third, and (d ) fourth moments for each of the sets of data (normal, skewed-right, skewed-left, uniform).

```python
def moments(data):
    mean = np.mean(data) # 1st moment
    std_dev = stats.stdev(data) # 2nd moment
    skewness =  pd.Series(data).skew() #3rd moment
    kurtosis = pd.Series(data).kurt() # 4th moment
    print(f"1st moment: {mean:.3f}")
    print(f"2nd moment: {std_dev:.3f}")
    print(f"3rd moment: {skewness:.3f}")
    print(f"4th moment: {kurtosis:.3f}")

print("Normal Data Set (female height measurements):")
moments(normal_data)
print()

print("Skewed-Right Data Set (age at marriage for females): ")
moments(skewed_right_data)
print()

print("Skewed-Left Data Set (obituary data that give the age at death for females):")
moments(skewed_left_data)
print()

print("Uniform Data Set (the amount of cola put into a 12 ounce container by a soft drinks machine):")
moments(uniform_data)
```

### Find the (a) first, (b) second, (c) third, and (d ) fourth moments about the mean for each of the sets of data (normal, skewed-right, skewed-left, uniform).

### Using the results of items 2 and 3 for the set of female height measurements, verify the relations between the moments

```python
# moments about the mean
def moments_mean(data):
    mean = np.mean(data)
    N = len(data)
    m1 = np.sum(data - mean) / N
    m2 = np.sum((data - mean) ** 2) / N 
    m3 = np.sum((data - mean) ** 3) / N 
    m4 = np.sum((data - mean) ** 4) / N 
    return m1, m2, m3, m4

# moments about the number 75 for the set of female height measurements
def moments_about_value(data, value):
    data = np.array(data)
    N = len(data)
    m1 = np.sum(data - value) / N
    m2 = np.sum((data - value) ** 2) / N
    m3 = np.sum((data - value) ** 3) / N
    m4 = np.sum((data - value) ** 4) / N
    return m1, m2, m3, m4

def verify_moments(data, reference_value):
    # Calculate moments about the mean (central moments)
    central_m1, central_m2, central_m3, central_m4 = moments_mean(data)
    
    # Calculate moments about the reference value (75 in this case)
    raw_m1, raw_m2, raw_m3, raw_m4 = moments_about_value(data, reference_value)
    
    print("Verifying relations between moments:")
    
    # Verify m2 = m2' - m1'^2
    m2_verify = raw_m2 - raw_m1**2
    print(f"m2 (calculated): {central_m2:.5f}")
    print(f"m2 (from raw moments): {m2_verify:.5f}")
    print(f"m2 verification result: {'Pass' if np.isclose(central_m2, m2_verify) else 'Fail'}")
    print()
    
    # Verify m3 = m3' - 3*m1'*m2' + 2*m1'^3
    m3_verify = raw_m3 - 3 * raw_m1 * raw_m2 + 2 * (raw_m1**3)
    print(f"m3 (calculated): {central_m3:.5f}")
    print(f"m3 (from raw moments): {m3_verify:.5f}")
    print(f"m3 verification result: {'Pass' if np.isclose(central_m3, m3_verify) else 'Fail'}")
    print()
    
    # Verify m4 = m4' - 4*m1'*m3' + 6*m1'^2*m2' - 3*m1'^4
    m4_verify = raw_m4 - 4 * raw_m1 * raw_m3 + 6 * (raw_m1**2) * raw_m2 - 3 * (raw_m1**4)
    print(f"m4 (calculated): {central_m4:.5f}")
    print(f"m4 (from raw moments): {m4_verify:.5f}")
    print(f"m4 verification result: {'Pass' if np.isclose(central_m4, m4_verify) else 'Fail'}")
    print()

# Normal Data Set (female height measurements) using reference value 75
print("Verifying moments for Normal Data Set (female height measurements) about 75:")
verify_moments(normal_data, 75)
```
